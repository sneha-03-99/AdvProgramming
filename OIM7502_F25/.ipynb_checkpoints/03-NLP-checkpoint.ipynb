```python
from bs4 import BeautifulSoup as bs # no, the pun is not lost 
import requests
from wordcloud import WordCloud, STOPWORDS
```

```python
url ="https://alphabench.com/data/python-loan-analysis.html"
```

```python
# get raw html
req=requests.get(url).text
req
```

```python
# get the first paragraph, with and without tag
#esc(changes), b- row below and a- row above 
soup= bs(req,'html.parser')
y=soup.p
y.text
```

```python
soup.find_all('p')
```

```python
# get the text of all hyperlinks
soup.a
```

```python
x=soup.find_all('a')
x
```

```python
for link in soup.find_all('a'):
    print(link.text)
```

```python
# get the URLs for all hyperlinks
for link in x:
    print(link.get('href'))
```

```python
# get text in paragraphs
text=''
for ps in soup.find_all('p'):
    text += ps.text
text
```

```python
text.split()
#edge case - we're
```

```python
# tokenize text from webpage
#regex101.com
from nltk.tokenize import RegexpTokenizer as Reg
tkz=Reg(r"(w+'?\w*)")
```

```python
text=tkz.tokenize(text)
```

```python
import nltk
text="".join([word for word in text if word not in nltk.corpus.stopwords.words('english')])
```

```python
from nltk import ngrams, FreqDist
freqs=dict()
for ngrams in 1,2,3,4:
    freqs[ngrams]= FreqDist(ngrams(text.split(), ngrams))
freqs[2]
```
