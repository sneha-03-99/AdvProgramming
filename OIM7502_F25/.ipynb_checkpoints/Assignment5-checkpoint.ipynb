```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

df = pd.read_csv('data/homes.csv')
df = df.sort_values(by='Sell', ascending=True)
df.head()
```

We begin by importing the pandas library and reading the homes.csv file into a DataFrame called df. The head() function shows the first five rows, giving an initial view of the dataset.

```python
df.info()
```

The info() function displays the column names, data types, and non-null values â€” helping us understand the structure and completeness of the dataset.



```python
df.describe()
```

The describe() function provides key statistics such as mean, median, and standard deviation for each numeric variable, giving insights into their distributions.

```python
df.isnull().sum()
```

Checking for missing values ensures data completeness. If any columns have many missing values, we may need to handle them before analysis.

```python
df = df.dropna()
```

To handle the missing values we use the dropna() function. It removes every row that has atleast one NaN(missing) value in any column. 

```python
plt.hist(df['Sell'], bins=20, color='skyblue', edgecolor='black', rwidth=0.9)
plt.title('Distribution of Home Selling Prices')
plt.xlabel('Selling Price')
plt.ylabel('Frequency')
plt.xticks(plt.xticks()[0][::2],rotation=45)
plt.tight_layout()
plt.show()
```

The histogram shows how selling prices are distributed â€” whether they are normally distributed or skewed. This helps understand data spread and potential outliers.

```python
plt.figure(figsize=(10, 8))
plt.scatter(df['Taxes'], df['Sell'], color='orange')
plt.title('Relationship between Taxes and Selling Price')
plt.xlabel('Taxes')
plt.ylabel('Selling Price')
plt.tight_layout()
plt.show()
```

We use the LinearRegression model from sklearn to predict Sell based on Taxes.
	

```python
X = df[['Taxes']]  # independent variable
y = df['Sell'] 
model = LinearRegression()
model.fit(X, y)
print("Intercept:", model.intercept_)
print("Coefficient for Taxes:", model.coef_[0])
print("R-squared:", model.score(X, y))
```

    â€¢Intercept represents the baseline selling price when taxes are zero.
	â€¢	Coefficient shows how much selling price increases for every unit increase in taxes.
	â€¢	R-squared measures how well the model explains the variation in selling price (closer to 1 means a better fit).
    If RÂ² is low, taxes alone may not be a strong predictor of selling price.

```python
plt.figure(figsize=(10, 8))
plt.scatter(df['Acres'], df['Sell'], color='red')
plt.title('Relationship between Acres and Selling Price')
plt.xlabel('Acres')
plt.ylabel('Selling Price')
plt.tight_layout()
plt.show()

X1 = df[['Acres']]  # independent variable
y1 = df['Sell'] 
model = LinearRegression()
model.fit(X1, y1)
print("Intercept:", model.intercept_)
print("Coefficient for Acres:", model.coef_[0])
print("R-squared:", model.score(X1, y1))
```

When comparing the two models, based on teh R-squared coefficients we can understand that taxes and Selling price have a higher correlation whereas the Acres and selling price have a weaker relationship. Also for the first one its easier to see a clear upward line through the scatter plot which isnt the same for the second plot. 

	â€¢	Simple linear regression only measures the impact of one variable at a time.
	â€¢	Real-world house prices depend on many interacting factors such as neighborhood, size, and age.
	â€¢	Outliers can distort the regression line.
	â€¢	Correlation â‰  causation â€” just because taxes correlate with price doesnâ€™t mean one causes the other
